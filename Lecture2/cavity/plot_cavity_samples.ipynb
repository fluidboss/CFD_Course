This notebook loads `U` and `p` sampled along the vertical and horizontal centre lines of an OpenFOAM v2312 cavity calculation. Run the following command in the case before using the notebook:

```bash
postProcess -func sampleDict -fields '(U p)' -latestTime
```

````python
from pathlib import Path
import sys
import matplotlib.pyplot as plt

# Make the companion Python module importable when the notebook is
# opened from this example directory.
sys.path.insert(0, str(Path.cwd()))
from plot_cavity_samples import load_cavity_samples, plot_cavity_samples
````

## Select the OpenFOAM case

If this notebook is not inside the cavity case, replace `.` with the full path to the case directory.

````python
CASE_DIRECTORY = Path('.')       # Example: Path('/home/user/OpenFOAM/cavity')
CAVITY_LENGTH = 0.1              # metres

print('Case:', CASE_DIRECTORY.resolve())
````

````python
time_directory, vertical, horizontal = load_cavity_samples(CASE_DIRECTORY)

print('Loaded time:', time_directory.name)
print('Vertical points:', len(vertical.distance))
print('Horizontal points:', len(horizontal.distance))
````

````python
fig, axes = plot_cavity_samples(
    vertical,
    horizontal,
    cavity_length=CAVITY_LENGTH,
    title=f'Cavity-flow centre lines — time {time_directory.name}',
)
plt.show()
````

````python
output_file = Path('cavity_profiles.png')
fig.savefig(output_file, dpi=200, bbox_inches='tight')
print('Saved:', output_file.resolve())
````

> **Pressure note:** For incompressible OpenFOAM solvers, `p` is normally kinematic pressure (`p/ρ`) in m²/s². If you require pressure in pascals, multiply the sampled values by the fluid density.
